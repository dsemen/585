```python
# functions 

def ispath(link):
    """Returns True if path is correct"""
    return Path(link).exists()


def cor_link(tov):
    """checks if file is in folder  and returns correct link"""
    link_if = '//gold585.int/uk/Общее хранилище файлов/Коммерческий департамент/Отдел закупки/КАТАЛОГИ/Nav/ИФ/'
    link_bk = '//gold585.int/uk/Общее хранилище файлов/Коммерческий департамент/Отдел закупки/КАТАЛОГИ/Nav/БК/'
    link_germes = '//gold585.int/uk/Общее хранилище файлов/Коммерческий департамент/Отдел закупки/КАТАЛОГИ/ГЕРМЕС/'
    link_germ_new = '//gold585.int/uk/Общее хранилище файлов/Коммерческий департамент/Гермес/'
    
    pic = tov + '.jpg'
    pic2 = tov + '#ТД.jpg'
       
    if ispath(link_if + pic) == True:
        return link_if + pic
    
    elif ispath(link_bk + pic) == True:
        return link_bk + pic
    
    elif ispath(link_germes + pic) == True:
        return link_germes + pic
    
    elif ispath(link_germes + pic2) == True:
        return link_germes + pic2
    
    elif ispath(link_germ_new + pic) == True:
        return link_germ_new + pic
    
    elif ispath(link_germ_new + pic2) == True:
        return link_germ_new + pic2
    
    else:
        return "no link"
```

```python
# check link with if/bk and returns correct link or true/false for current link

path = r'\\gold585.int\uk\Общее хранилище файлов\Коммерческий департамент\Отдел закупки\ЛИЧНЫЕ\Семен\4.xlsx'

df = pd.read_excel(path, usecols=[0,1])

df = df.iloc[:33]

df['Copy_to'] = df['арт'].apply(lambda x: cor_link(x))

df.to_excel('C:\Остатки\неправильные_фото.xlsx', index=False)

```
