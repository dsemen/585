##### Формула для модели сигнала: s(t)=sin(2πθ1t)*exp(-t/θ2), 5/T<θ1<15/Т, 3T/2<θ2<5Т/2 (где Т=1). Распределение аддитивной помехи – Лапласовское (двойное экспоненциальное).
## Решение

```python
import numpy as np
import math as m

import matplotlib.pylab as plt

pi = m.pi
x = np.linspace(1, 20, num=1000)
q1 = np.linspace(5, 15, num=1000)
q2 = np.linspace(1.5, 2.5, num=1000)

pdf = np.sin(2 * pi * q1 * x) * np.exp(-x / q2)
pdf2 = pdf + np.random.laplace(0., 0.5, size=1000)

fig = plt.figure(figsize=(15, 10))
plt.ylabel('s(t)')
plt.xlabel('t')
plt.plot(x, pdf)
plt.title(r'$s(t)=s(t) =\sin(2 \pi  \theta_1 t) \times \exp( \frac{-t}{\theta_2})$'+
             '\n при '+r'$\frac{5}{T} < \theta_1 < \frac{15}{T}$' + ' и ' + r'$\frac{3}{2}T < \theta_2 < \frac{5}{2}T$',
             fontsize = 20, pad = 20)
plt.show()
#plt.plot(x, pdf1)
```

```python
import matplotlib.pyplot as plt
import numpy as np


a1=5.1
a2=1.51

loc, scale = 0., 0.5

x = np.linspace(-1.5*np.pi, 1.5*np.pi, 1000)
y = lambda x: np.sin(2*np.pi*a1*x)*np.exp(-x/a2)
pomeha = np.random.laplace(loc, scale, 1000)

plot = plt
plot.plot(x, y(x)+pomeha)
# Plot.title('example')
plot.ylabel('yp(x)')
plot.xlabel('x')
plot.legend(['yp(x)'], loc='upper right')
plot.show()


plot2 = plt
count, bins, ignored = plt.hist(pomeha, 30, density=True)
xxx = np.arange(-4., 4., .01)
pdf = np.exp(-abs(xxx-loc)/scale)/(2.*scale)
plot2.plot(xxx, pdf)

g = (1/(scale * np.sqrt(2 * np.pi)) *
     np.exp(-(xxx - loc)**2 / (2 * scale**2)))
plot2.plot(xxx,g)
plot2.show()
```
